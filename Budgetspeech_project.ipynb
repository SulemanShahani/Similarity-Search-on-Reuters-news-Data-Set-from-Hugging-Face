```python
# Importing components from LangChain for vector store interaction with Cassandra



#Importing Hugging Face's Transformers library
from transformers import AutoTokenizer, AutoModel
import torch
# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("BAAI/bge-small-en-v1.5")
model = AutoModel.from_pretrained("BAAI/bge-small-en-v1.5")


# Set the padding token ID in the model's configuration
model.config.pad_token_id = tokenizer.pad_token_id

# Adjust tokenizer maximum sequence length to match model configuration
tokenizer.model_max_length = model.config.max_position_embeddings
tokenizer.padding_side = "left"





# Loading datasets using Hugging Face library
from datasets import load_dataset

from PyPDF2 import PdfReader

import numpy as np

import torch
```

```python
# provide the path of  pdf file/files.
pdfreader = PdfReader('Budget_Speech_english_2022_23.pdf')
```

```python

raw_text_pages = []  # List to store content from each page separately

for i, page in enumerate(pdfreader.pages):
    content = page.extract_text()
    if content:
        raw_text_pages.append(content)  # Append content of each page to the list

# If you want to concatenate all the pages into a single string:
raw_text = '\n'.join(raw_text_pages)  # Join content of all pages with newline characters

# Clean the text
raw_text = raw_text.strip()  # Remove leading and trailing whitespace
raw_text = ' '.join(raw_text.split())  # Remove extra whitespace


import re  # Import the re module for regular expressions

import string
# Define a translation table to remove punctuation
# Clean the text
raw_text = raw_text.strip()  # Remove leading and trailing whitespace
raw_text = re.sub(r'[^\w\s\.,!?]', '', raw_text)  # Remove symbols except punctuation


# Print or inspect the cleaned text
print("Cleaned Text:")
print(raw_text)
```

```python


# Function to split text into smaller paragraphs or sentences
import re

def split_text_into_paragraphs(text):
    # Split text into paragraphs based on the pattern "number + period"
    paragraphs = re.split(r'\d+\.', text)
    # Remove empty paragraphs and leading/trailing whitespace
    paragraphs = [p.strip() for p in paragraphs if p.strip()]
    return paragraphs



# Split the raw text into paragraphs
splitted_text = split_text_into_paragraphs(raw_text)


paragraphs = splitted_text[0:100]

# Define a function to find the maximum paragraph length
def max_paragraph_length(paragraphs):
    max_length = 0
    for paragraph in paragraphs:
        # Split the paragraph into words
        words = paragraph.split()
        # Get the length of the paragraph (number of words)
        paragraph_length = len(words)
        # Update the maximum length if needed
        if paragraph_length > max_length:
            max_length = paragraph_length
    return max_length

# Find the maximum paragraph length
max_length = max_paragraph_length(paragraphs)


# Add a padding token to the tokenizer
tokenizer.pad_token = tokenizer.eos_token
# Process each paragraph through the model



import torch

# Define a function to generate embeddings for paragraphs and apply padding
def generate_embeddings_with_padding(paragraphs, model, tokenizer, max_length):
    # Initialize an empty list to store the padded embeddings
    padded_embeddings = []

    # Iterate through each paragraph
    for paragraph in paragraphs:
        # Tokenize the paragraph
        inputs = tokenizer(paragraph, return_tensors="pt", padding="max_length", max_length=512, truncation=True)

        # Generate embeddings from the model
        with torch.no_grad():
            outputs = model(**inputs)

            embeddings = outputs.last_hidden_state.mean(dim=1)

        # Append the embeddings to the list
        padded_embeddings.append(embeddings)

    return padded_embeddings






# Add a padding token to the tokenizer
tokenizer.pad_token = "[PAD]"



# Define the maximum paragraph length
max_paragraph_length = 512  # Update with your maximum paragraph length

# Generate embeddings for paragraphs and apply padding
padded_embeddings = generate_embeddings_with_padding(paragraphs, model, tokenizer, max_paragraph_length)


```

```python








import faiss

class FaissIndexWrapper:
    def __init__(self, index):
        self.index = index

    def add(self, vectors):
        self.index.add(vectors)

    def search(self, query_vector, k):
        distances, indices = self.index.search(query_vector, k)
        return distances, indices

# Convert padded embeddings to a numpy array
padded_embeddings_np = torch.cat(padded_embeddings).numpy()

# Define the dimensionality of the embeddings
dimensionality = padded_embeddings[0].shape[-1]

# Create a FAISS index
index = faiss.IndexFlatL2(dimensionality)

# Add vectors to the index
index.add(padded_embeddings_np)

# Create an instance of FaissIndexWrapper
faiss_index_wrapper = FaissIndexWrapper(index)
```

```python
def generate_query_embedding(query_text):
    # Tokenize the query text data
    querytokenized_input = tokenizer(query_text, return_tensors="pt", padding=True, truncation=True)

    # Pass the tokenized input through the model
    with torch.no_grad():
        outputs = model(**querytokenized_input)

    # Extract the embeddings from the model outputs
    queryembeddings = outputs.last_hidden_state[:, 0]  # Extract [CLS] token embedding from the last layer

    queryembeddings_array = queryembeddings.numpy()

    return queryembeddings_array


def interact():
    while True:
        # Prompt the user for a question
        query_text = input("Enter your question (or type 'quit' to exit): ")

        # Check if the user wants to quit
        if query_text.lower() == "quit":
            print("Exiting...")
            break

        # Convert query text to embedding
        query_embedding = generate_query_embedding(query_text)

        # Perform similarity search
        k = 1  # Number of nearest neighbors to retrieve
        distances, indices = index.search(query_embedding,  k)

        # Retrieve text results corresponding to the nearest neighbors' indices
        nearest_neighbors_text = [paragraphs[i] for i in indices[0]]



        # Print similar documents and their distances
        print("Similar documents:")
        for distance, document in zip(distances[0], nearest_neighbors_text):
            print(f"Distance: {distance}, Document: {document}")


# Run the interaction loop
interact()
```
